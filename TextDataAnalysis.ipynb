### Finding Ngram 
N-grams are contiguous sequences of N elements, such as words or letters, retrieved from a given text in Python NLP. They are used to record and analyse language trends in the local setting. N-grams can be constructed in Python using tools such as NLTK by tokenizing the text and constructing subsequences of N elements. N-grams are useful for applications such as language modelling, information retrieval, and machine translation because they provide insight into the structure and relationships of a text.

Nithyashree V (2021). What Are N-Grams and How to Implement Them in Python? [online] Analytics Vidhya. Available at: https://www.analyticsvidhya.com/blog/2021/09/what-are-n-grams-and-how-to-implement-them-in-python/ [Accessed 5 Jul. 2023].

‌

```python
# This function aims to clean up the data by performing several steps. First, 
# it identifies and removes all words that are not considered stop words. 
# Next, the remaining words are encoded and undergo basic regex parsing. 
# Finally, the lemmatization process is applied to the encoded and parsed words.

# De, C. (2020). From DataFrame to N-Grams - Towards Data Science. 
# [online] Medium. Available at: https://towardsdatascience.com/from-dataframe-to-n-grams-e34e29df3460 [Accessed 5 Jul. 2023].

# GeeksforGeeks. (2018). Python Lemmatization with NLTK.
#  [online] Available at: https://www.geeksforgeeks.org/python-lemmatization-with-nltk/ [Accessed 5 Jul. 2023].
def basic_clean(text):
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english')
  text = (unicodedata.normalize('NFKD', text)
    .encode('ascii', 'ignore')
    .decode('utf-8', 'ignore')
    .lower())
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

**Word2vec is a neural network consisting of two layers that works by transforming text into vector representations of words, commonly known as word embeddings. It takes a text corpus as input and generates a collection of vectors as output. These vectors serve as feature representations, effectively converting words into a numerical format that can be comprehended by deep neural networks. Although Word2vec itself is not considered a deep neural network, it plays a crucial role in translating textual data into a format that deep neural networks can process and analyze effectively (Dhruvil Karani, 2018).**

Dhruvil Karani (2018). Introduction to Word Embedding and Word2Vec - Towards Data Science. [online] Medium. Available at: https://towardsdatascience.com/introduction-to-word-embedding-and-word2vec-652d0c2060fa [Accessed 5 Jul. 2023].


‌

```python
y = news_['news_class'].values
X = []
stop_words = set(nltk.corpus.stopwords.words("english"))
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
for par in news_["news"].values:
    tmp = []
    sentences = nltk.sent_tokenize(par)
    for sent in sentences:
        sent = sent.lower()
        tokens = tokenizer.tokenize(sent)
        filtered_words = [w.strip() for w in tokens if w not in stop_words and len(w) > 1]
        tmp.extend(filtered_words)
    X.append(tmp)
```

```python
EMBEDDING_DIM = 100
w2v_model = gensim.models.Word2Vec(sentences=X, vector_size=EMBEDDING_DIM, window=5, min_count=1)
```

### Modeling and Training to detect Fake News<br>
![traingif](https://thumbs.gfycat.com/CommonCircularIrishwaterspaniel-size_restricted.gif)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(news_['news'], news_['news_class'], random_state=42)
```

```python
# The training data is fitted to the CountVectorizer
vect = CountVectorizer().fit(X_train)
```

```python
X_train_vectorized = vect.transform(X_train)

X_train_vectorized
```

```python
from sklearn.linear_model import LogisticRegression
# Train the model
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import roc_auc_score
# Predict the transformed test documents
predictions = model.predict(vect.transform(X_test))
print('AUC (Area Under The Curve): ', roc_auc_score(y_test, predictions))
```

The AUC (Area Under the Curve) of this model is 0.9984, indicating that it is an excellent fit. The high AUC score suggests that the model has a remarkable ability to accurately distinguish between positive and negative instances. With such a strong AUC score, we can confidently conclude that the model performs exceptionally well in its classification task.

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, predictions, labels=[0, 1]))
# Printing the precision and recall, among other metrics
print(metrics.classification_report(y_test, predictions, labels=[0, 1]))
```

## Latent Dirichlet Allocation

```python
# Machinelearninggeek.com. (2021). Latent Dirichlet Allocation using Scikit-learn – Machine Learning Geek. 
# [online] Available at: https://machinelearninggeek.com/latent-dirichlet-allocation-using-scikit-learn/ [Accessed 5 Jul. 2023].
news_list= newsClean2['text'].tolist()
```

```python
# Initialize regex tokenizer
tokenizer = RegexpTokenizer(r'\w+')

# Vectorize using TF-IDF
tfidf = TfidfVectorizer(lowercase=True,
                        stop_words='english',
                        ngram_range = (1,1),
                        tokenizer = tokenizer.tokenize)

# Fit & Transform the documents
train_data = tfidf.fit_transform(news_list)   
```

```python
# Define the number of topics
num_components=7
model=LatentDirichletAllocation(n_components=num_components)
lda_matrix = model.fit_transform(train_data)
lda_components=model.components_
```

```python
# Print the topics with their terms
terms = tfidf.get_feature_names()
for index, component in enumerate(lda_components):
    zipped = zip(terms, component)
    top_terms_key=sorted(zipped, key = lambda t: t[1], reverse=True)[:7]
    top_terms_list=list(dict(top_terms_key).keys())
    print("TOPIC "+str(index)+": ",top_terms_list)
```

### The LDA analysis produced illuminating results, demonstrating the presence of seven separate topic groups. Each topic group is made up of terms that are closely related and provide useful insights into the contents and themes of the respective news stories. One instance example is subject 3, which includes keywords such as 'trump', 'said', 'president', 'clinton', 'house', 'people', and 'obama'. These keywords are primarily related to presidents, indicating that the news articles associated with this topic are most likely focused on political news, including discussions, statements, and activities involving notable political figures such as Trump, Clinton, Obama, and White House matters. We acquire a better grasp of the underlying themes and subjects covered in the news dataset by evaluating these topic groupings and their associated keywords.
